```python
import pandas as pd
tables = [f"MULTPL/{region}" for region in pd.DataFrame(ds.to_list())[1].to_list()]
tables.sort()
tables
```

```python
data_SP500 = []

for region in tables:
  df_tmp = nasdaqdatalink.get(region, start_date="1999-12-31", end_date="2022-12-31", paginate=True)
  df_tmp.columns = [region.split('/')[-1]]
  if 'SP500_REAL_PRICE_YEAR' in region:
    data_SP500.append(df_tmp)
```

```python
pd.concat(data_SP500).sort_index().ffill().plot(figsize=(13, 10), subplots=True)
```
